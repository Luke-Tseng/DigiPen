```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from datetime import datetime
from datetime import timedelta
import time
from IPython import display
%matplotlib inline
```

#### Load datasets in pandas dataframe


```python
apple = pd.read_csv('AAPL.csv')
google = pd.read_csv('GOOG.csv')
microsoft = pd.read_csv('MSFT.csv')
```

#### Plot open prices as time series

```python
apple['Date'] = pd.to_datetime(apple['Date'])
microsoft['Date'] = pd.to_datetime(microsoft['Date'])
google['Date'] = pd.to_datetime(google['Date'])

plt.figure(figsize=(10,5))

plt.plot(apple['Date'], apple['Open'], label='AAPL', color='red')
plt.plot(google['Date'], google['Open'], label='GOOG', color='blue')
plt.plot(microsoft['Date'], microsoft['Open'], label='MSFT', color='green')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Open Stock Prices')

plt.legend()
plt.tight_layout()
plt.show()
```

All 3 stocks have an upward trend over the years, but Google has the highest increase over the years, going from the 200s into almost the 3000s.

I would buy the Google stock, maybe around 2020 since that's when the price of the stock drops the most.

```python
# Get microsoft as stock with 2018 as year
microsoft_2018 = microsoft[microsoft['Date'].dt.year == 2018]

# Calculate mean
microsoft_2018['Mean'] = (microsoft_2018['High'] + microsoft_2018['Low']) / 2

# Plotting high and low with stack plot and mean with line plot
plt.figure(figsize=(10, 5))
plt.stackplot(microsoft_2018['Date'], microsoft_2018['Low'], microsoft_2018['High'] - microsoft_2018['Low'], labels=['Low', 'High'], colors=['lightblue', 'red'])
plt.plot(microsoft_2018['Date'], microsoft_2018['Mean'], label='Mean', color='black')

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('MSFT Stock Prices in 2018')
plt.legend()
plt.show()
```

#### Plotting rolling window means

```python
plt.figure(figsize=(10,5))

# weekly = 5, monthly = 20, yearly = 250

# Apple
plt.plot(apple['Date'], apple['Open'].rolling(5).mean(), label='Apple Weekly', color='red', linestyle='solid')
plt.plot(apple['Date'], apple['Open'].rolling(20).mean(), label='Apple Monthly', color='red', linestyle='dotted')
plt.plot(apple['Date'], apple['Open'].rolling(250).mean(), label='Apple Yearly', color='red', linestyle='dashed')

# Google
plt.plot(google['Date'], google['Open'].rolling(5).mean(), label='Google Weekly', color='blue', linestyle='solid')
plt.plot(google['Date'], google['Open'].rolling(20).mean(), label='Google Monthly', color='blue', linestyle='dotted')
plt.plot(google['Date'], google['Open'].rolling(250).mean(), label='Google Yearly', color='blue', linestyle='dashed')

# Microsoft
plt.plot(microsoft['Date'], microsoft['Open'].rolling(5).mean(), label='Microsoft Weekly', color='green', linestyle='solid')
plt.plot(microsoft['Date'], microsoft['Open'].rolling(20).mean(), label='Microsoft Monthly', color='green', linestyle='dotted')
plt.plot(microsoft['Date'], microsoft['Open'].rolling(250).mean(), label='Microsoft Yearly', color='green', linestyle='dashed')

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Rolling Means')
plt.legend()
plt.show()
```

I would buy the Google stock as early as possible, since the stock is continuously increasing and increases the most compared to the others
